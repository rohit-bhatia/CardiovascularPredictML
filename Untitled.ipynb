# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
actual_pred = model.predict(x_train)
```

```python
print(accuracy_score(y_test,y_pred))
#print(f1_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_train,actual_pred))
print(f1_score(y_train,actual_pred))
print(confusion_matrix(y_train,actual_pred))
print(classification_report(y_train,actual_pred))
```

# XGboost

```python
from xgboost import XGBClassifier
from xgboost import plot_importance
```

```python
model = XGBClassifier(random_state=2019,n_estimators=500,learning_rate=0.05,max_depth=2,colsample_bytree=0.7,subsample=0.7)
```

```python
model.fit(x_train,y_train,eval_set=[(x_train,y_train),(x_test,y_test)],verbose=False,early_stopping_rounds=50)
```

```python
y_pred = model.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(f1_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
plot_importance(model)
```

```python
results = model.evals_result()
```

```python
iterations = len(results['validation_0']['error'])
```

```python
plt.figure(figsize=(10,10))
plt.plot(range(iterations),results['validation_0']['error'],label="training error")
plt.plot(range(iterations),results['validation_1']['error'],label="validation error",c="orange")
plt.legend()
```
